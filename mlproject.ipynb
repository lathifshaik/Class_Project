# Logistic Regression

```python
def getLRmodel():
    input_layer = Input(shape=(None,22))
    dense_layer_1 = Dense(256,activation='relu')(input_layer)        
    dense_layer_2 = Dense(128,activation='relu')(dense_layer_1)
    dense_layer_3 = Dense(64,activation='relu')(dense_layer_2)
    dense_layer_4 = Dense(32,activation='relu')(dense_layer_3)
    output_1 = Dense(1,activation='sigmoid',name="max_temp")(dense_layer_4)
    output_2 = Dense(1,activation='sigmoid',name="min_temp")(dense_layer_4)
    logisticRegression = Model(input_layer,[output_1,output_2])
    logisticRegression.compile(
        optimizer='adam',
        loss='mean_squared_error',
        metrics=[tf.keras.metrics.RootMeanSquaredError()])
#     print(logisticRegression.summary())
    return logisticRegression
test = getLRmodel()
```

```python
tf.keras.utils.plot_model(test, "multi_input_and_output_model.png")
```

```python
model_02 = getLRmodel()
h = model_02.fit(X_train,y_train,epochs=50)
```

```python
plt.plot(h.history['max_temp_root_mean_squared_error'])
plt.plot(h.history['min_temp_root_mean_squared_error'])
plt.show()
```
